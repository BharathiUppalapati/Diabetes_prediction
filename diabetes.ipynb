```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
h=OneHotEncoder()
h=['Pregnancies','Glucose','BloodPressure','SkinThickness'
,'Insulin','BMI','DiabetesPedigreeFunction','Age']
df=pd.get_dummies(data,columns=h)
df
```

```python
le=LabelEncoder()
df['Outcome']=le.fit_transform(df['Outcome'])
df
```

```python
x=df.drop(['Outcome'],axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=42)
print(x_train)
```

```python
y_train
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
db=GaussianNB()
db.fit(x_train,y_train)
```

```python
predictions=db.predict(x_test)
predictions
```

```python
ac=accuracy_score(y_test,predictions)
con=confusion_matrix(y_test,predictions)
cl=classification_report(y_test,predictions)
print(ac)
print(con)
print(cl)
```
